```python
import googlemaps

def get_coordinates(address):
    # Google Maps API 키를 여기에 입력하세요
    gmaps = googlemaps.Client(key='AIzaSyDClh8EDTMFdOqTqtLcLXcHy8Hm-BvuowQ')

    
    # 지오코딩 요청
    geocode_result = gmaps.geocode(address)
    
    if geocode_result:
        lat = geocode_result[0]['geometry']['location']['lat']
        lng = geocode_result[0]['geometry']['location']['lng']
        return lat, lng
    else:
        return None
```

```python
#성남시 행정동 GeoJSON 파일 추출
# 기존 GeoJSON 파일 불러오기
import json


geojson_path = "data/seongnam_geo.json"

with open(geojson_path, encoding="utf-8") as f:
    geo_json = json.load(f)

# GeoJSON 데이터에서 '경기도 성남시' 제거하고 '구별'과 '동별'을 분리
for feature in geo_json["features"]:
    original_name = feature["properties"]["adm_nm"]  # 예: "경기도 성남시수정구 신흥1동"
    
    # '경기도 성남시' 제거
    cleaned_name = original_name.replace("경기도 성남시", "").strip()  # "수정구 신흥1동"
    
    # 숫자로 끝나는 "1동", "2동", "3동" 등을 제거하여 기본 동이름만 사용 (신흥1동 → 신흥동)
    cleaned_name = ''.join([char for char in cleaned_name if not char.isdigit()]).replace("  ", " ").strip()
    
    # 새로운 속성 추가 (구, 동 나누기)
    gu_name, dong_name = cleaned_name.split(" ", 1)  # "수정구 신흥동" -> "수정구", "신흥동"
    feature["properties"]["gu_name"] = gu_name
    feature["properties"]["dong_name"] = dong_name

# 수정된 GeoJSON을 새로운 파일로 저장
new_geojson_path = "data/seongnam_geo.json"
with open(new_geojson_path, "w", encoding="utf-8") as f:
    json.dump(geo_json, f, ensure_ascii=False, indent=4)
```
